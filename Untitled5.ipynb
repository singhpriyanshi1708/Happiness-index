```
df_=df.drop(['Country or region'],axis=1)
```

```
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```
X=df_.drop(['Score'],axis=1)
y=df_['Score']
```

```
b=0
c=0
for i in range(42,102):
  ran=i
  X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=ran)
  lr=LinearRegression()
  lr.fit(X_train,y_train)
  pred=lr.predict(X_test)
  a=r2_score(y_test,pred)
  print(f'r2 score of random state of {ran} is {a}')

  if a>b:
    b=a
    c=ran
    print(f' max r2 score of random state of {c} is {b}')
```

```
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=57)
lr=LinearRegression()
lr.fit(X_train,y_train)
pred=lr.predict(X_test)
r2_score(y_test,pred)
 
```
